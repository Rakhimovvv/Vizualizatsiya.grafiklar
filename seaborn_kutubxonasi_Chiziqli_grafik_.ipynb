```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.plot( df)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.title("Number of museum visitors")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.ylabel("Number of visitors")
sns.lineplot(data=df  ["Chinese American Museum"])
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.title("Number of museum visitors")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.ylabel("Number of visitors")
sns.lineplot(data=df[["Chinese American Museum" ,"Firehouse Museum"]])
plt.show()
```
