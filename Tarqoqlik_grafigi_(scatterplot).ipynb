```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
plt.figure(figsize=(10, 6))
df =df[df.year>2000]
#sns.scatterplot(data=df, x="year", y="price")
sns.regplot(data=df, x="year", y="price")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
df =df[df.year>2000]
sns.scatterplot(data=df, x="year", y="price" , hue="transmission" , style="fuelType" , s=100)
#sns.regplot(data=df, x="year", y="price")
plt.show()
```

```python
sns.swarmplot(data=df, x="year", y="price")
plt.show()
```
